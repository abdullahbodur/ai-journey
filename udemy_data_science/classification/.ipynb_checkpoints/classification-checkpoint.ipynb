### Rok Eğrisi Bakma
Modelimizin başarı oranının grafiğidir. <br>
Amaç eşik ile grafiğin arasında kalan alanı maximize etmektir. <br>

***ROC*** : Receiver Operating Characteristics <br>
***AUC*** : Area under the curve <br>
***TPR*** : True Positive Recall <br>
***FPR*** : False Positive Recall <br>


```python
logit_roc_auc = roc_auc_score(y,loj_model.predict(X))

fpr , tpr , thresholds = roc_curve(y, loj_model.predict_proba(X)[:,1])

plt.figure()
plt.plot(fpr,tpr,label = 'AUC (area = %0.2f)' % logit_roc_auc)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('False Positive Oranı')
plt.ylabel('True Positive Oranı')
plt.title('ROC')
plt.show();
```

## Destek Vektör Makineleri (SVM)
Amaç iki sınıf arasındaki ayrımın optimum olmasını sağlayacak hiper-düzlemi bulmaktır. <br>


* marjin alanları (gri) alanların ***ayrıldığı*** ve ***noktaların içerilmediği*** ***en büyük doğru*** iç alanıdır. 
* turuncu noktalar ***destek*** noktalarıdır ve ***marjin bu noktalara sınırdır***, marjini bu noktalar belirler <br>

 <br>
Bu durumlarda boyut arttırma (Kernel Trick) yöntemi geliştirilmiştir. (***Sağ***) <br>
***mü (μ) ve sigma (σ)*** değerlerini doğru seçilerek bu ayırma en doğru şekilde yapılır. 

***Hiper parametreler***: <br>
* C

```python
diabetes = pd.read_csv('diabetes.csv')
df = diabetes.copy()
df = df.dropna()
y = df["Outcome"]
X = df.drop(['Outcome'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30, 
                                                    random_state=42)
```

```python
mlpc.coefs_
```

### Model Tuning
* activation ***'identity', 'logistic', 'tanh', 'relu'***, default='relu')
* solver ***'lbfgs', 'sgd', 'adam'*** default='adam' The solver for weight optimization.)
* alpha (L2 penalty (regularization term) parameter, default = 0.0001 )
* learning_rate ***'constant', 'invscaling', 'adaptive'*** default='constant' Learning rate schedule for weight updates.)

```python
mlpc_params = {"alpha": [0.1, 0.01, 0.02, 0.005, 0.0001,0.00001],
              "hidden_layer_sizes": [(10,10,10),
                                     (100,100,100),
                                     (100,100),
                                     (3,5), 
                                     (5, 3)],
              "solver" : ["lbfgs","adam","sgd"],
              "activation": ["relu","logistic"]}
```

```python
mlpc_cv_model = GridSearchCV(mlpc,mlpc_params,cv = 10,n_jobs = -1 , verbose = 2)
```

```python
mlpc_cv_model.fit(X_train_scaled,y_train)
```

```python
mlpc_cv_model.best_params_
```

```python
# tuned model
```

```python
tuned_mlcp = MLPClassifier(activation='logistic',hidden_layer_sizes = (100,100,100),solver = 'adam',alpha = 0.1)
```

```python
tuned_mlcp_model = tuned_mlcp.fit(X_train_scaled,y_train)
```

```python
# score
```

```python
y_pred = tuned_mlcp_model.predict(X_test_scaled)
```

```python
accuracy_score(y_test,y_pred)
```

## Random Forest Classification
 <br>

* Bagging (Breiman, 1996) ile Random Subspace (Ho, 1998) yöntemlerinin birleşimi ile oluşmuştur.

* Ağaçlar için gözlemler bootstrap rastgele örnek seçim yöntemi ile ***değişkenler random subspace yöntemi ile seçilir.***

* Karar ağacının her bir düğümünde en iyi dallara ayırıcı (bilgi kazancı) ***değişken tüm değişkenler arasından rastgele seçilen daha az sayıdaki değişken arasından seçilir.***

* Ağaç oluşturmada veri setinin .***2/3'ü.*** kullanılır. Dışarıda kalan veri ağaçların performans değerlendirmesi ve değişken öneminin belirlenmesi için kullanılır.

### Model & Test

```python
diabetes = pd.read_csv('diabetes.csv')
df = diabetes.copy()
df = df.dropna()
y = df["Outcome"]
X = df.drop(['Outcome'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30, 
                                                    random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier().fit(X_train,y_train)
```

```python
# test - score
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

### Model Tuning
* ***n_estimators*** : ağaç (tahmin) sayısı * birincil önem
* ***max_features*** : bölünmelerde dikkate alınacak max değişken sayısı * birincil önem

* max_depth : ağacın max derinliği (derinlik yükseldikçe daha çok bölünme olur.) 
* min_samples_split : her bir bölünme için en az gerekli olan örnek sayısı
* min_samples_leaf : her bir yapraktaki min örnek sayısı

```python
rf_params = {'max_depth' :[2,5,8,10] , 
             'max_features' : [2,5,8], # 8 tane değişken olduğundan 8 durumu begging durumudur. 
             'n_estimators' : [10,500,1000],
             'min_samples_split' : [2,5,10]
            } 
```

```python
rf_model = RandomForestClassifier()
```

```python
rf_cv_model = GridSearchCV(rf_model,rf_params, cv = 10, n_jobs = -1, verbose = 2)
```

```python
rf_cv_model.fit(X_train,y_train)
```

```python
# best_params
```

```python
rf_cv_model.best_params_
```

```python
# tuned model
```

```python
rf_tuned = RandomForestClassifier(max_depth = 10,
                                  max_features=8,
                                  min_samples_split=2,
                                  n_estimators = 500)
```

```python
# predict - score
```

```python
y_pred = rf_tuned.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

### değişken önem düzeyi

```python
Importance = pd.DataFrame({'Importance' : rf_tuned.feature_importances_ * 100},index = X_train.columns)
```

```python
Importance.sort_values(by = 'Importance',axis = 0,ascending  = True).plot(kind = 'barh',color = 'r')

plt.xlabel('Değişken Önem Düzeyleri')
```
