### Rok Erisi Bakma
Modelimizin baar orannn grafiidir. <br>
Ama eik ile grafiin arasnda kalan alan maximize etmektir. <br>

***ROC*** : Receiver Operating Characteristics <br>
***AUC*** : Area under the curve <br>
***TPR*** : True Positive Recall <br>
***FPR*** : False Positive Recall <br>


```python
logit_roc_auc = roc_auc_score(y,loj_model.predict(X))

fpr , tpr , thresholds = roc_curve(y, loj_model.predict_proba(X)[:,1])

plt.figure()
plt.plot(fpr,tpr,label = 'AUC (area = %0.2f)' % logit_roc_auc)
plt.plot([0,1],[0,1],'r--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('False Positive Oran')
plt.ylabel('True Positive Oran')
plt.title('ROC')
plt.show();
```

```python
mlpc.coefs_
```

### Model Tuning
* activation ***'identity', 'logistic', 'tanh', 'relu'***, default='relu')
* solver ***'lbfgs', 'sgd', 'adam'*** default='adam' The solver for weight optimization.)
* alpha (L2 penalty (regularization term) parameter, default = 0.0001 )
* learning_rate ***'constant', 'invscaling', 'adaptive'*** default='constant' Learning rate schedule for weight updates.)

```python
mlpc_params = {"alpha": [0.1, 0.01, 0.02, 0.005, 0.0001,0.00001],
              "hidden_layer_sizes": [(10,10,10),
                                     (100,100,100),
                                     (100,100),
                                     (3,5), 
                                     (5, 3)],
              "solver" : ["lbfgs","adam","sgd"],
              "activation": ["relu","logistic"]}
```

```python
mlpc = MLPClassifier()
```

```python
mlpc_cv_model = GridSearchCV(mlpc,mlpc_params,cv = 10,n_jobs = -1 , verbose = 2)
```

```python
mlpc_cv_model.fit(X_train_scaled,y_train)
```

```python
mlpc_cv_model.best_params_
```

```python
# tuned model
```

```python
tuned_mlcp = MLPClassifier(activation='logistic',hidden_layer_sizes = (100,100,100),solver = 'adam',alpha = 0.1)
```

```python
tuned_mlcp_model = tuned_mlcp.fit(X_train_scaled,y_train)
```

```python
# score
```

```python
y_pred = tuned_mlcp_model.predict(X_test_scaled)
```

```python
accuracy_score(y_test,y_pred)
```

## Random Forest Classification
 <br>

* Bagging (Breiman, 1996) ile Random Subspace (Ho, 1998) yntemlerinin birleimi ile olumutur.

* Aalar iin gzlemler bootstrap rastgele rnek seim yntemi ile ***deikenler random subspace yntemi ile seilir.***

* Karar aacnn her bir dmnde en iyi dallara ayrc (bilgi kazanc) ***deiken tm deikenler arasndan rastgele seilen daha az saydaki deiken arasndan seilir.***
.***
* Aa oluturmada veri setinin .***2/3'.*** kullanlr. Darda kalan veri aalarn performans deerlendirmesi ve deiken neminin belirlenmesi iin kullanlr.

* Her dm noktasnda rastgele deiken seimi yaplr. .***(regresyon'da p/3, snflama'da karekk p).***

* Nihai tahmin iin aalardan tahmin deerleri talep edilirken her bir aacn daha nce hesaplanan hata oranlar gz nne alnarak .***aalara arlk verilir..***
