### Bernoulli Dağılımı ###
Başarılı-başarısız, olumlu-olumsuz şeklindeki iki sonuçlu olaylar
ile ilgilenildiğinde kullanılan kesikli olasılık dağılımıdır.

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$


${\displaystyle \operatorname {E} \left(X\right)=p}$

$\operatorname {Var} [X]=pq=p(1-p)$

```python
from scipy.stats import bernoulli
```

```python
# yazı tura olasılığı => tura gelme olasılığı 0.6
p = 0.6
```

```python
rv = bernoulli(p)
rv.pmf(k=1) # probability mass functions (pmf), 

# k=1 gerçekleşmesini istediğimiz olay.
# k=0 ikinci olay.
```

```python
rv.pmf(k=0)
```

## Büyük Sayılar Yasası

**Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.**

Amaç deney sayısını arttırarak söylenen olaslığı gözlemlemeye başlıyoruz

```python
import numpy as np
rng = np.random.RandomState(123)
for i in np.arange(1,21):
    try_count = 2**i
    heads_and_tails = rng.randint(0,2,size=try_count)
    tail_prob = np.mean(heads_and_tails)
    print('Atış Sayısı :',try_count,'---',' yazı sayısı :%.2f ' % (tail_prob))
```

## Binom Dağılımı

${\displaystyle f(k;n,p)={n \choose k}p^{k}(1-p)^{n-k}}$, k = 0, 1, 2, ...,n


${\displaystyle \operatorname {E} (X)=np}$


${\displaystyle \operatorname {Var} (X)=np(1-p)}$

### Binom Dağılımı Grafiği

### Benzerlikler

n = 4 , p=0.5, x=2




## Binom Problem

**Problem:**
Çeşitli mecralara reklam veriliyor, reklamların tıklanma ve geri dönüşüm <br>
oranları optimize edilmeye çalışılıyor. Buna yönelik olarak belirli bir <br>
mecrada çeşitli senaryolara göre reklama tıklama olasılıkları <br>
hesaplanmak isteniliyor.


**Detaylar:**
Bir mecrada reklam verilecek
Dağılım ve reklama tıklama olasılığı biliniyor (0.01)


**Soru: Reklamı 100 kişi qördüğünde 1,5,10 tıklanması olasılığı nedir?**



```python
 from scipy.stats import binom
```

```python
p = 0.01 # tıklama olasılığı
n = 100 # kaç deneme
rv = binom(n,p)
print('100 kişi arasından 1 kişinin tıklama olasılığı %.2f' % (rv.pmf(1)*100))
print('100 kişi arasından 5 kişinin tıklama olasılığı %.2f' % (rv.pmf(5)*100))
print('100 kişi arasından 10 kişinin tıklama olasılığı %.10f' % (rv.pmf(10)*100))
```

## Poisson Dağılımı

${\displaystyle f(x,\lambda )={\frac {\lambda ^{x}e^{-\lambda }}{x!}}\,\!}$,  x = 0,1,2,...,n

### Grafik


### Poisson Dağılımına Örnekler:
* 10 BİN kelimeden oluşan bir kitapta hatalı kelime sayısı
* 4000 öğrencili okulda not girişinde hata yapılması

* Rötara düşen uçuş sefer sayısı

### Kurallar

n > 50 => gözlem sayısı <br>
n*p = λ & n*p < 5 <br>

* Rassal denemeler iki sonuçlu olmalı
* Aynı koşullar altında gerçekleştirilmelidir
* Rassal denemeler birbirinden bağımsız olmalıdır

### Poisson Soru :
Bir üniversitede 5000 not girişinde 5 tane notun yanlış <br>
girilmesi olasılığı nedir? <br>

## Poisson Problem 

***Problem:***
Hatalı ilan girişi olasılıkları hesaplanmak isteniyor. <br>
***Detaylar:***
Bir yıl süresince ölçümler yapılıyor <br>
Dağılım biliniyor (Poisson) ve Lambda 0.1 (ortalama hata sayısı) <br>

Hiç hata olmaması, 3 hata olması ve 5 hata olması olasılıkları nedir? <br>


```python
from scipy.stats import poisson
```

```python
lambda_ = 0.1
rv = poisson(mu=lambda_)
```

```python
print('hiç hatalı giriş olmama olasılığı %.2f' % (rv.pmf(0)*100))
print('3 hatalı giriş olma olasılığı %.4f' % (rv.pmf(3)*100))
print('3 hatalı giriş olma olasılığı %.7f' % (rv.pmf(5)*100))
```

## Normal Dağılım ##
Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesaplaması için kullanılır.

${\displaystyle f(x\mid \mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}}$

* $\mu$  ortalama ya da dağılımın beklenen değeri

* ${\displaystyle \sigma }$  standart sapma

* ${\displaystyle \sigma ^{2}}$ varyans

kesikli rassal değişkenlerde x değerlerimiz ***sabit*** olduğundan formül direkt bize sonucu veriyor idi. <br>
sürekli rassal değişkenlerde bizden aralığın olasılığını soracaklarından ***integral*** ile sonucu bulabiliriz.

***Problem:*** <br>
Bir yatırım/toplantı öncesinde gelecek ay ile ilgili satışların <br>
belirli değerlerde gerçekleşmesi olasılıkları belirlenmek isteniyor. <br>
***Detaylar:*** <br>
Dağılımın normal olduğu biliniyor <br>
Aylık ortalama satış sayısı 80K, standart sapması 5K <br>

90K'dan fazla satış yapma olasılığı nedir? <br>


```python
from scipy.stats import norm
```

```python
# eski fonksiyonlarda kesikli olduğundan => probability mass func (pmf)
# sürekli olasılık olduğunda ise => cumulative density func
1 - norm.cdf(90,80,5)

# cdf fonksiyonu cumulative olduğundan 0-90 arasını hesaplar. biz de 1 den çıkararak geri kalan yeri hesapladık.
```

```python
# 70'den fazla olması
1 - norm.cdf(70,80,5)
```

```python
# 73'den az olması
norm.cdf(73,80,5)
```

```python
# 85-90 arasında olması
norm.cdf(90,80,5) -  norm.cdf(85,80,5)
```

### Hipotez Nedir? ###
Bir inanışı (bir savı, bir tahmini vs) test etmek için kullanılan istatistiksel bir tekniktir. <br>

Örn bir doktor bulduğu bir ilacın bir tür hastalıkta iyi geldiğini ispatlaması için hipotez testi yapmalıdır. <br>
sonucunda gelen istatistiğe göre bu ilacın o hastalığa iyi geldiği savı ispatlanmış olur.

### Hipotez ve Türleri ###
2 tip'dir:
* ***hipotez***
* ***alternatif-hipotez***



## Hata Tipleri ##



h0 gerçek değerini bilemeyiz. Bu sebep ile biz bilinmeyen durumda hipotezi reddederek veya reddetmeyerek ilerlenir. <br>

* h0 gerçekte doğru (hiçbir zaman bilemeyiz) biz de reddetmessek (+) => 1-α(alpha)
* h0 gerçekte doğru (hiçbir zaman bilemeyiz) biz de reddetersek  (-) => α(alpha)
* h0 gerçekte yanlış (hiçbir zaman bilemeyiz) biz de reddetmessek (+) => β(beta)
* h0 gerçekte yanlış (hiçbir zaman bilemeyiz) biz de reddetersek  (-) => 1-β(alpha)

## P-Value
Hipotez testlerinin sonuçlarını değerlendirmek üzere programlar tarafından p-value değeri verilir. <br>
Bu değer üzerinden kolayca yorum yapabiliriz.

İstatistik camiasında α(alpha) değeri ***0.05*** olarak sabitlenmiştir. Eğer bizim p-value değerimiz <br>
bu değerden ***küçük*** ise h0 hipotezini ***reddederek*** kolayca çözümleyebiliriz. <br>

Ama örnek dağılımın dağılımı ile teorik dağılımın dağılımı ile birbirine eşittir hipotezini uygulamak için <br>
reddetmek istemeyiz. Çünkü kullandığımız normallik varsayımı gibi varsayımları farklı şekilde test ederiz.

### Hipotez Testi Adımları

***Adım 1***: Hipotezlerin kurulması ve yönlerinin belirlenmesi 





***Adım 2:*** Anlamlılık düzeyinin ve tablo değerinin belirlenmesi<br> 
* *α değeri 0.05 olarak kabul edilir*<br>

***Adım 3:*** Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması<br>

***Adım 4:*** Hesaplanan test istatistiği ile alfa'ya karşılık gelen tablo
değerinin karşılaştırılması.<br>
Test İstatistiği (Zh) > Tablo Değeri (Zt) ise H0 Red

***Adım 5:*** yorum

## Tek Örneklem T Testi

Populasyon ortalaması ile varsayımsal bir değer arasında istatistiksel olarak anlamlı <br>
bir farklılık olup olmadığını test etmek için kullanılan parametrik bir testtir.

Örnek verecek olursak 10bin kişilik bir populasyonun içinden bir örneklem çekdiğimizi düşünürsek

***Tek Örneklem*** => Çekilen Örneklem 

yapacağımız iş örneklem üzerinden ortalama hesaplamak olursa ***Tek Örneklem T Testi*** denir <br>

### Test İstatistiği
 <br>

1. Anakütle standart sapması biliniyorsa z
istatistiği kullanılır.
2. Anakütle standart sapması bilinmiyorsa
ve n > 30 ise z istatistiği kullanılır.
3. Anakütle standart sapması bilinmiyor ve
n < 30 ise t istatistiği kullanılır.

***n büyüdükçe t, z'ye yaklaşır*** <br>
Örneklerimiz sayısı (n) genellikle 30 dan büyük olacağından t varsayabiliriz

### Varsayım
son olarakta hipotez içerisindeki varsayımları test etmek gerekir. <br>
eğer parametrelerimiz yok ise non-parametrik, var ise parametrik varsayımları uygulamalıyız. <br>
güncel hayatta genel olarak non-parametrik varsayımlar vardır.

## İş Uygulaması : Web Sitesinde Geçirilen Sürenin Testi
 <br>
***Problem:*** <br>
Web sitemizde geçirilen ortalama süre gerçekten 170 saniye mi?<br>
***Detaylar:***
Yazılımlardan elde edilen web sitesinde geçirilen ort. süreler var. Bu veriler incelendiğinde <br>
bir yönetici ya da çalışanımız bu değerlerin böyle olmadığına yönelik düşünceler taşıyor ve<br>
bu durumu test etmek istiyorlar.<br>

Örnek ortalamasına ilişkin test yapmak için kullanılır.

**Hipotezler:**

<d><i>H</i><sub>0</sub>: <i>μ</i> = <i>170</i></d>   
<d><i>H</i><sub>1</sub>: <i>μ</i> ≠ <i>170</i></d><br><br>


<table></table>

**Test İstatistiği:**

${\displaystyle t={\frac {{\bar {x}}-\mu _{0}}{\frac {s}{\sqrt {n}}}}}$

```python
import numpy as np

calculations = np.array([17, 160, 234, 149, 145, 107, 197, 75, 201, 225, 211, 119, 
              157, 145, 127, 244, 163, 114, 145,  65, 112, 185, 202, 146,
              203, 224, 203, 114, 188, 156, 187, 154, 177, 95, 165, 50, 110, 
              216, 138, 151, 166, 135, 155, 84, 251, 173, 131, 207, 121, 120])
```

```python
import scipy.stats as stats
```

```python
# betimsel istatistiklerine bakalım. Bu sefer scipy içerisindeki bir fonksiyonu kullanacağız.
stats.describe(calculations)
#skewness => basıklık , kurtosis => çarpıklık
```

Mean 154 geldi. Peki bize verilen hipotezde 170 olması soruluyordu. Ortalaması 154 olan şeyin 170 olur mu?  <br>

* Ya 154 örneklemden kaynaklı yanlış bir sonuç ise (şansa bağlı)
* Amacımız bir savın doğruluğunu 95% doğruluk ile tespit etmek

### Dağılımın Varsayımını bulma
* Normallik Varsayımız
    - Grafik Üzerinden bulma (***yanılmaya elverişli***)
        * histogram üzerinden
        * qqplot üzerinden
        

```python
import pandas as pd
```

```python
# histogram
pd.DataFrame(calculations).plot.hist();
```

```python
# qqplot
import pylab
stats.probplot(calculations,dist='norm',plot=pylab)
pylab.show()

# y-axis => örnekteki gerçek değerler
# x-axis => örneğin teorik dağılım değerleri

# amaç doğrunun etrafında dağılım gösterdiğini görebilmek. 
# Görülebiliyor ise normal dağılımdır.
```

```python
# Shapiro-Wilks Testi => dağılımın normal dağılım olduğu hipotezini kontrol eden test.

# amacımız neydi bir verinin ortalamasının belirli bir değer etrafında olduğunu ispatlamak.

# bu hipotezi ben bulurum fakat önce senin verdiğin veri kümesinin dağılımının normal olduğunu görebilmemiz gerek.
```

H0 => örnek dağılım ile teorik normal dağılım arasında istatistiksel olarak anlamlı bir farklılık yoktur (***Normal Dağılım***)<br> 
H1 => .... fark vardır.

```python
from scipy.stats import shapiro 
```

```python
shapiro(calculations)
```

***p-value > α*** olduğundan H0 hipotezi reddedilemez. Yani ***dağılım normaldir***. 

```python
# varsayım sağlandı (Normallik Varsayımı)
```

### Hipotez Testinin Uygulaması

```python
stats.ttest_1samp(calculations,popmean=170)
```

***p-value < α*** olduğundan H0 ***Hipotezi Reddedilir***

Yani ortalama 170 arasındadır hipotezi ***yanlıştır***. <br>

## Nonparametrik Tek Örneklem Testi
Bundan önceki yaptığımız test ***parametrik test*** idi. Parametrik olunca örneklem herhangi bir varsayım tipini içeriyor <br>
olmalıydı. Bundan önceki örnekte ***Normal Varsayım*** gözlemlemiştik. Fakat güncel hayat problemlerinde bir varsayıma <br>
bağlamak (***parametrik***) zor olmaktadır. Bu sebeple güncel hayat problemlerinde ***Nonparametrik*** (varsayımsal dağılım içermeyen) <br>
yapılar gözlemleyeceğiz. Peki varsayım içermeyen örneklemlerde hiptez testi nasıl yapılır?

```python
# üstteki shapiro testi p-value < α çıktı ve H0'ı reddettik. Bu durumda non-parametrik hipotez uygulayacağız.
```

```python
from statsmodels.stats.descriptivestats import sign_test 
```

```python
sign_test(calculations,170) # işaret testi
```

***p-value > α*** => H0'ı ***reddedemiyoruz***. ama bizim shapiro değerimiz düzgün çıktığından normal dağılıma sahibiz. <br>
yani bu sonuca bakmayacağız.

## Tek Örneklem Oran Testi : Hipotezler
Populasyonun bir verisi olan a/b değeri için çekilen örneklem üzerinden elde edilir.

***Örnekler***<br>

### Tek Örneklem Oran Testi : Problem 

Problem:
Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından ***0.125 dönüşüm oranı*** elde <br>
edildiği ifade edilmiş. Fakat bu durum kontrol edilmek isteniyor. Çünkü bu ***yüksek bir oran*** ve gelirler <br>
incelendiğinde ***örtüşmüyor.*** <br>

***Detaylar:*** <br>
500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alişveriş yapmış. <br>
Örnek üzerinden elde edilen dönüşüm oranı: 40/500 = 0,08 <br>



```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40 # 40'kere alışveriş yapılmış
nobs = 500 # gözlem sayısı
value = 0.125 # test edilecek değer
```

```python
proportions_ztest(count,nobs,value) # p-value < 0.05 => h0 hipotezi red
```

```python
shapiro(ABC.group_c) # p-value > a => ✅
```

```python
from scipy.stats import levene
```

```python
levene(ABC.group_a,ABC.group_b,ABC.group_c) # p-value > a => ✅
```

### Varsayım Sonucu
* gruplar bağımsız ✅
* normallik => ✅
* varyans homojenliği => ✅

```python
f_oneway(ABC.group_a,ABC.group_b,ABC.group_c) # p-value < a => ❌
```

## Korelasyon Analizi

Değişkenler arasındaki ***ilişki***, bu ***ilişkinin yönü*** ve ***şiddeti*** ile ilgili bilgiler sağlayan istatistiksel bir yöntemdir. <br>
 <br>

### Korelasyonlara Örnekler
 


### Korelasyonun Anlamlılığının Testi
bundan önce korelasyon işlediğimizde anlamlılığı üzerine bir yorum yapmadık.<br>
Burada amacımız bu ilişki ***anlamlı bir ilişki*** mi yoksa ***tesadüfi*** mi ?

 <br>
#### p nedir? 
ρ, (Ro Sembolü) korelasyonun anlamlılığının ölçüsüdür. Korelasyonun ***anlamlı olabilmesi*** için ***H0'ı reddetmeliyiz.*** <br>
p = 0 => ***anlamlılık yoktur***.<br>
p!= 0 => ***anlamlılık vardır***.<br>

### Test İstatistiği


### Varsayımlar
* İki değişken içinde normallik varsayımı.
* Varsayım sağlanıyorsa ***Pearson Korelasyon Katsayısı***
* Varsayım sağlanmıyorsa ***Spearman Korelasyon Katsayısı***

###  Yorumlar
pearson korelasyon katsayısı (r) üzerinden (-1) - (1) arasında değer alınarak ilişkinin yönü ile ilgili bilgi alırız. <br>
anlamlılık testi, yani hipotez ile de bu ilişkinin tesadüfi olarak anlamlı olup olmadığını buluruz.


### İş Uygulaması : Bahşiş İle Hesap Arasındaki İlişkinin İncelenmesi 
Bahşiş ile ödenen hesap arasında korelasyon var mı ?  <br>
 <br>

Bahşiş veri seti:

total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)

tip: bahşiş

sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)

smoker: grupta sigara içen var mı? (0=No, 1=Yes)

day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)

time: ne zaman? (0=Day, 1=Night)

size: grupta kaç kişi var?


```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
# total_bill içerisinde bahşiş de olduğundan bu veriyi total_bill den çıkarmamız gerek
```

```python
df['total_bill'] = df.total_bill - df.tip
df.head()
```

```python
# basit bir şekilde korelasyon grafiğine bakalım. 
df.plot.scatter('tip','total_bill');
# görüldüğü üzere total bill arttıkça tips de artmış
```

### Varsayım Kontrolü 

```python
# normallik varsayımı
```

```python
from scipy.stats import shapiro
```

```python
shapiro(df.total_bill) # p-value < a => ❌
```

```python
shapiro(df.tip) # p-value < a => ❌
```

***Normallik şartı sağlanmıyor***, bu durumda bizim ***nonparametrik*** yola başvurmamız gerek. <br>
Fakat göstermelik parametrik testi de kullanacağız

## Hipotez Testi

### Korelasyon Katsayısı
***corr*** methodu default olarak ***pearson*** methodunu kullanır. <br>
Bizim varsayımlarımız negatif çıktığından ***spearman*** kullanmamız lazım.

* ***varsayım pozitif ise (+) => pearson***
* ***varsayım negatif ise (-) => spearman***

```python
df['tip'].corr(df.total_bill) 
```

```python
df['tip'].corr(df.total_bill,method='spearman') # katsayı pozitif yönlü çıktı. ama bu sayı anlamlı mı bakalım.
```

## Korelasyon Anlamlılığının Testi

### Parametrik Test (Pearson)

```python
from scipy.stats.stats import pearsonr
```

```python
pearsonr(df.tip,df.total_bill) # p-value < a , ❌ # H0 reddedildi => anlamlı bir ilişki vardır.
```

### Nonparametrik Test (Spearman)

```python
from scipy import stats
```

```python
stats.spearmanr(df.tip,df.total_bill) # p-value < a , ❌ # H0 reddedildi => anlamlı bir ilişki vardır.
```

```python
 # bir farklı algoritma da => kendalltau
```

```python
stats.kendalltau(df.tip,df.total_bill) # p-value < a , ❌ # H0 reddedildi => anlamlı bir ilişki vardır.
```
