## K-En Yakın Komşu (KNN)
tahminler gözlem benzerliğine göre yapılır. // verilen veriye en yakın tahmini verir <br>
Bana arkadaşını söyle sana kim olduğunu söyleyelim. <br>
Başta ***classification için*** çıkmış. Sonrada ***regression***'a uygulanmıştır <br>

### Nasıl Çalışır
 <br>

### KNN Basamakları
* Komşu sayısını belirle (K)
* Bilinmeyen nokta ile diğer tüm noktalar ile arasındaki uzaklıkları hesapla
* Uzaklıkları sırala ve belirlenen k sayısına göre en yakın olan k gözlemi seç
* Sınıflandırma ise en sık sınıf, regresyon ise ortalama değeri tahmin değeri olarak ver.

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

### Model oluşturma

```python
knn_model = KNeighborsRegressor().fit(X_train,y_train)
```

```python
# knn hiperparametresi
```

```python
knn_model.n_neighbors
```

### Tahmin

```python
y_pred = knn_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# farklı k (n_neighbors) değerlerine karşı hata değeri
```

```python
RMSE = []

for k in range(10):
    k = k+1
    knn_model = KNeighborsRegressor(n_neighbors = k).fit(X_train,y_train)
    y_pred = knn_model.predict(X_train)
    rmse = np.sqrt(mean_squared_error(y_train,y_pred))
    RMSE.append(rmse)
    print(f'k = {k} için RMSE değeri : {rmse}')
```

### Model Tuning

```python
from sklearn.model_selection import GridSearchCV
```

### Grid Search Çalışma Mantığı

duruma bağlı n adet hiperparametrenin birbirlerine göre ***oluşabilecek tüm kombinasyonlarını*** deneyen bir algoritmadır. <br>
denerken CV kullanır. <br>

```python
knn_params = {'n_neighbors' : np.arange(1,30,1)} # parametre adlarını doğru yazmak önemli 
```

```python
knn = KNeighborsRegressor()
```

```python
knn_cv_model = GridSearchCV(knn,knn_params,cv=10)
```

```python
knn_cv_model.fit(X_train,y_train)
```

```python
## en uygun parametre ne
```

```python
knn_cv_model.best_params_['n_neighbors']
```

```python
## yukarıdaki CV olmadan ölçüm ve CV Ölçüm arasındaki fark
```

```python
RMSE = []
RMSE_CV = []

for k in range(10):
    k = k+1
    knn_model = KNeighborsRegressor(n_neighbors = k).fit(X_train,y_train)
    y_pred = knn_model.predict(X_train)
    rmse = np.sqrt(mean_squared_error(y_train,y_pred))
    rmse_cv = np.sqrt(-cross_val_score(knn_model, X_train,y_train , cv=10 , scoring = 'neg_mean_squared_error')).mean()
    
    RMSE.append(rmse)
    RMSE_CV.append(rmse_cv)
    
    print(f'k = {k} , için RMSE değeri : {rmse}, RMSE_CV değeri : {rmse_cv}')
```

```python
## görüldüğü üzere başta aldığımız değerlerdeki standart sapma yüksek iken daha stabil sonuçlar aldık
```

```python
## RMSE ye göre en optimize sonuc k=2 (179 RMSE) | RMSE_CV göre en optimize k=8 (RMSE_CV = 280)
```

```python
## bu sebep ile RMSE_CV kullanmak daha optimize sonuç almamızı sağlar
```

```python
## valide edilmiş durum daha iyidir.
```

### Tuned edilmiş model oluşumu

```python
knn_tuned = KNeighborsRegressor(n_neighbors =knn_cv_model.best_params_['n_neighbors'] )
```

```python
knn_tuned.fit(X_train,y_train)
```

```python
y_pred = knn_tuned.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

## Destek Vektör Regresyonu (SVR)
Amaç, bir marjin aralığına maksimum noktayı en küçük hata ile alabilecek şekilde doğru ya da eğriyi belirlemektir. <br>
Çıkış olarak ***classification için çıkmış*** sonra da regresyonlara uygulanmıştır. <br>

Yapısı itibari ile ***Robust Regresyon*** olarak da anılmaktadır. <br>

***Robust Regresyon Nedir?*** <br>
Linear Regresyonda ortaya çıkan problemlere (çok boyutluluk - aykırı gözlemlere duyarlılık vs) dayanıklı modellere denir. <br>

### Formülizasyon 
 <br>

***Formülizasyon Açıklaması***<br>
* w değeri bizim ß1 değerimizdir. 
* C değeri bizim ceza parametremizdir.
* ksi(ξ) değerleri grafiğin kabuk kısmının(ε) dışında olan noktaların kabuk kısmına(ε) olan uzaklıklarıdır.


***Kısıtlar*** <br>
Öyle bir grafik çiz ki bütün noktalar grafiğe epsilon(ε) + ksi(ξ) ' değerinden uzak olamasın. <br>

### Önemli Notlar
* ***ξ değerleri mutlak haldedir.***
* ***grafiğin eğimini kabuk dışında kalan değerler (aykırı gözlem) belirlemektedir.***
* ***kabuk içindeki değerler grafiğin ta kendisidir.***
* ***kabuk dışında kalan değerleri yönetebilmek için ceza işlemi uygulanmaktadır.***

## Doğrusal Olmayan SVR

```python
# el ile bir veri seti oluşumu
```

```python
np.random.seed(3)
```

```python
x_sim = np.random.uniform(2, 10, 145)
y_sim = np.sin(x_sim) + np.random.normal(0, 0.4, 145)
```

```python
x_outliers = np.arange(2.5, 5, 0.5)
y_outliers = -5*np.ones(5)
```

```python
x_sim_idx = np.argsort(np.concatenate([x_sim, x_outliers]))
x_sim = np.concatenate([x_sim, x_outliers])[x_sim_idx]
y_sim = np.concatenate([y_sim, y_outliers])[x_sim_idx]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ols = LinearRegression()
ols.fit(np.sin(x_sim[:, np.newaxis]), y_sim)
ols_pred = ols.predict(np.sin(x_sim[:, np.newaxis]))
```

```python
from sklearn.svm import SVR
```

```python
eps = 0.1
svr = SVR('rbf', epsilon = eps) # radial basis func
svr.fit(x_sim[:, np.newaxis], y_sim)
svr_pred = svr.predict(x_sim[:, np.newaxis])
```

```python
plt.scatter(x_sim,y_sim,alpha=0.5,s=26)
plt_ols = plt.plot(x_sim,ols_pred,'g')
plt_svr = plt.plot(x_sim,svr_pred,'r')
plt.xlabel('Bagımsız Degisken')
plt.ylabel('Bagımlı Degisken')
plt.ylim(-5.2 , 2.2)
plt.legend([plt_ols,plt_svr] ,['EKK',]);
```

```python
# görüldüğü üzere yeşil grafik linear_regression (EKK) üzerine kurulu olduğundan aykırı gözlemlerden kaynaklı birkaç yerde sapmalar olmuş
```

## Yapay Sinir Ağları
İnsan beyninin bilgi işleme şeklini referans alan sınıflandırma ve regresyon <br>
problemleri için kullanılabilen kuvvetli makine öğrenmesi algoritmalarından <br>
birisidir. <br> 

 <br>

* Dentritler bir önceki sinir hücresinden gelen bilgiyi alır. 
* Her bir Dentritin kendine özel bir ağırlığı vardır.
* Bu sebep ile çekirdeğe gelen bilgi bu ağırlıklar hesaba katılarak işleme alınır.
* Soma bütün sinyalleri alır
* Akson Somadan bilgileri alır ve Sinapsislere taşır.
* Sinapsislerde diğer hücreye gönderilmeden önce bir ön işlem yapılır.
* Ön işlem ile sinyaller karşı tarafın eşik değerine uydurulur.

 <br>

* Normal regresyondan farklılaştığı bölge ***Aktivasyon Fonksiyonudur***.
* Burada değişkenler çarpımından çıkan sonuç işleme tabi tutulur. 

 <br>

***Çalışma Basamakları***
* Ağ içine inputlar verilir.
* Rastgele başlangıç ağırlıklar oluşturulur.
* Katman içinde tahmini değer ve gerçek değer arasındaki fark hesaplanır.
* Eşik değere yakınlaşana dek ağırlıklar üzerinde düzenlemeler yapılır.
* Eşik değere ulaşınca sonuç çıktısı verilir.

 <br>

* k değeri ***gizli katman sayısını*** temsil eder.
* p ***ağırlık adedini*** belirtir. (feature adedi)
* ***her katmanda*** linear regresyondaki gibi ***her bir değişken bir ağırlık ile çarpılır***.
* ***g fonksiyonu*** gibi bir ***aktivasyon fonksiyonu*** ile (buradaki fonk sigmoid fonksiyonu) bir sonraki katmana hazırlanır.
* f fonksiyonu da birden çok katmanın birlikte çalışımını gösterir.
* en sonundaki algoritma da ElasticNet'e benzemektedir. Ceza uygulamaları vardır. 
* toplanacak olursa toplam tahmin edilecek katsayı sayısı = H*(p+1) + h+1 
* bu katsayıları MSE - SSR gibi fonksiyonlar ile ölçülür.
* Başta ilk katsayılar random atanır.

### Bazı bilgiler
* optimum parametre değerleri back propagation (geriye yayılım) ile bulunur. [türevlerle çalışır]
* back propagation'ın dezavantajı ise başka bir bulunan parametre setinden daha iyi olduğunun bir açıklaması yoktur.
* yapay sinir ağları overfit etmeye meğillidir.
* Engellemek için eğitimi durdurma yöntemleri geliştirilmiştir

### Modelleme

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
# elimizdeki veri setini bir standart aralığa sokmamız lazım.
# yapay sinir ağları variance yüksek olan yapılarda güvensizlik yaratır.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled  = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# modelimizi kuralım
```

```python
from sklearn.neural_network import MLPRegressor # Multilayer Perceptron (Çok katmanlı algılayıcı)
```

```python
mlp_model = MLPRegressor((100,20)).fit(X_train_scaled,y_train) # 100 , 20 hidden layerların katmanlardaki eleman sayıları
```

```python
# katman sayısı / input layer - hidden layer 1 - hidden layer2 - output layer
```

```python
mlp_model.n_layers_
```

```python
# gizli katmandaki eleman sayısı / 1 tane olduğundan sadece birinci katmanın sayısı var
```

```python
mlp_model.hidden_layer_sizes
```

### Tahmin

```python
y_pred = mlp_model.predict(X_test_scaled)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

### Model Tuning

```python
mlp_params = {'alpha' : [0.1,0.01,0.02,0.005], 'hidden_layer_sizes' : [(20,20),(100,50,150),(300,200,150)], 'activation' : ['logistic','relu']}
```

```python
mlp_cv_model = GridSearchCV(mlp_model,mlp_params,cv=10)
```

```python
mlp_cv_model.fit(X_train_scaled,y_train)
```

```python
mlp_cv_model.best_params_
```

## Karar Ağaçları (Classification and Regression Trees - CART)
Amaç veri seti içerisindeki karmaşık yapıları basit karar yapılarına dönüştürmektir. <br>
Heterojen veri setleri belirlenmiş bir hedef değişkene göre homojen alt gruplara ayrılır. <br>

* Ayırma işlemlerini chi-square (ki-kare) , cini , entropi algoritmaları yapar.
* Tahmin gücü düşüktür. Fakat ilerideki modellemelere temel olacaktır. 

 <br>
### Teknik Analiz
* bu görselde 2 tane değişken üzerinden karar ağacı oluşturulmuştur.
* görselde 2 tane iç düğüm (internal node) vardır.
* görselde 3 tane yaprak (terminal node) vardır.
* son yaprak 6.74 değeri olandır.
* yapraklar o bölgenin izdüşümünün aralıklarının ortalaması ile oluşturulur. 

### Analitik Analiz
* Ağacın değişkenlerine odaklanacak olursak, yukarıya doğru değişkenin açıklayıcılık (önem) değeri artar.
* Deneyim yılı en açıklayıcı değişkendir.
* Deneyim yılı 4.5 dan düşük ise atış sayısı maaşını etkilememektedir.

 <br>
Sağdaki görselin tamamını heterojen bir örneklem kümesi varsayar isek. Bunların belirli değişkenlere göre <br>
Dallara ayrılması ve küçük kümecikler oluşturulması sonucu içinde birkaç veya birçok örneklem bulunan gruplar oluşur. <br>
Bu grupların ortalaması alınarak bölgelerin eşik değerleri oluşur

* aşırı dallanma, overfitting var - traini çok iyi tahmin ediyor.
* kurucusunun önerdiği bir yöntem ile overfitting engellenebiliyor - Ceza İşlemi

### Nerede Kullanılmalı ?
Karar Ağaçları genellenebilirlik kaygısı olmayan yerlerde çok iyi sonuçlar verir.
* Örn spesifik bir departmanın verilerinden oluşturulmuş bir karar ağacı modeli ile o departmana gelecek yeni birisinin maaşı hesaplanabilir.
* Fakat tüm şirkete uygulanırsa başarılı sonuçlar almaz.
* Varyans / Varyasyon açıklamada başarılıdır.
* Rassallık var ise problem çıkarır.

## Bagging
 <br>
 <br>

* Bootstrap yöntemi ile seçilim yapılır.
* veri setinde m adet gözlem vardır
* n adet gözlem bootstrap (rastgele) ile seçilir. // yerine koymalı
* t adet karar ağacı oluşturulur
* çekilen örneklerin 2/3'ü ağaçların oluşturulması kalanı ise performans öçlümü için kullanılır.
* ezberlemeye dayanlıklı , MSE değerini düşüren, rastsallık problemini düzeltmiştir.

## Random Forest
Temeli birden çok karar ağacının ürettiği tahminlerin bir araya getirilerek <br>
değerlendirilmesine dayanır.

### Özellikler
* Bagging (Breiman, 1996) ile Random Subspace (Ho, 1998) yöntemlerinin birleşimi ile oluşmuştur.

* Ağaçlar için ***gözlemler bootstrap*** rastgele örnek seçim <br>
yöntemi ile ***değişkenler random subspace*** yöntemi ile seçilir.

* Karar ağacının ***her bir düğümünde*** en iyi dallara ayırıcı (bilgi kazancı) değişken,  <br>
tüm değişkenler arasından ***rastgele seçilen*** daha az sayıdaki değişken arasından seçilir. <br>
(***tüm değişken sayısı p; p>k; k adet değişken arasından düğüm seçilir***)

* ***Ağaç oluşturmada veri setinin 2/3'ü kullanılır.*** Dışarıda ***kalan*** veri ağaçların ***performans değerlendirmesi*** <br>
ve ***değişken öneminin belirlenmesi*** için kullanılır. 

* Her düğüm noktasında rastgele değişken seçimi yapılır. <br>
(***regresyon'da p/3, sınıflama'da karekök p***)

* Nihai tahmin için ağaçlardan tahmin değerleri talep edilirken her bir ağacın <br> ***daha önce hesaplanan hata oranları göz önüne alınarak ağaçlara ağırlık*** verilir.

### Model

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state = 42)
```

```python
rf_model.fit(X_train,y_train)
```

```python
# n_estimators => ağaç sayısı ***
# max_features => düğüm bölgelerinde bölünmelerde dikkate alıncak max özellik sayısı ***
# max_depth => her ağacın derinlik seviyesi / overfit için ***
# min_samples_split => her bölme noktasındaki min gözlem sayısı / overfit için
# min_samples_leaf => leaf_node'daki min gözlem sayısı / overfit için
```

### Tahmin

```python
y_pred = rf_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

### Model Tuning
2 önemli hiperparametre var <br>
* n_estimators => ağaç sayısı
* max_features => düğüm bölgelerinde dikkate alınacak max özellik sayısı 

***max_depth*** ( derinlik ayarlayıcı ) overfit için ayarlanabilir.

```python
rf_params = {'max_depth' : list(range(1,10)),
            'max_features' : [3,5,10,15],
            'n_estimators' : [200,500,1000,2000]} 

# ilgili makelelerde 500 / 1000 değerleri daha iyi çalışıldığını gözlemlenmiş 
```

## Gradient Boosting Machines
AdaBoost'un sınıflandırma ve regresyon problemlerine kolayca uyarlanabilen genelleştirilmiş versiyondur. <br>
Artıklar üzerine tek bir tahminsel model formunda olan modeller serisi kurulur <br>
 <br>
 <br>
* ***Bagging'de*** karar ağaçları, ***birbirinden bağımsızdır***.
* ***Boosting*** işlemlerinde karar ağaçları cumulative şekilde ***birleştirilir***. Ve ***güçlü bir karar ağacı*** oluşturulur. <br>
 <br>
* Box1, Box2, Box3 tek başlarına zayıf tahminler yapar. Bunların birleşiminden oluşan Box4 ise güçlü tahmin yapar.

### Gradient Boosting Machine
AdaBoost'un genelleştirilmiş ve regresyonlara uygulanabilen hali'dir.

### GBM Özellikleri
* Gradient boosting tek bir tahminsel model formunda olan modeller serisi oluşturur.

* Seri içerisindeki bir model serideki ***bir önceki modelin <br>tahmin artıklarının/ hatalarının*** (residuals) ***üzerine kurularak*** (fit) oluşturulur.

* GBM diferansiyellenebilen herhangi bir kayıp fonksiyonunu 
öptimize edebilen ***Gradient descent algoritmasını*** kullanmakta.

* GB bir çok temel öğrenici tipi (base learner type) kullanabilir. (***Trees, linear terms, splines,…***)

* ***Cost fonksiyonları ve link fonksiyonları modifiye edilebilirdir***.

* ***Boosting + Gradient Descent***

* GBM bu zamana kadar öğrendiğimiz tüm modellerin ölçeklendirilmiş framework'u olarak adlandırılabilir. <br> ***bundan sonraki modeller*** ise GBM'in ***performans arttırılmaları*** olarak ortaya çıkmıştır. <br>

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbm_model = GradientBoostingRegressor()
gbm_model.fit(X_train,y_train)
```

```python
# GBM hem karar ağacı modelini hem linear modeli destekler.
# Biz Burada karar ağac modelini dikkate alacağız.

# criterion (kriter) : friedman_mse
# max_depth : derinlik
# max_features : maximum özellik sayısı
# learning_rate : öğrenme oranı alpha / Gradient Descent kullanır.
# max_leaf_node : max-yaprak sayısı
# n_estimators: karar ağacı sayısı
```

### Tahmin

```python
y_pred = gbm_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred)) # ilkel hata
```
